```python

import pandas as pd

df = pd.read_csv('emails.csv')

print("Dataset Preview:")
print(df.head())

df.columns = df.columns.str.lower()
if 'label' in df.columns:
    label_col = 'label'
elif 'spam' in df.columns:
    label_col = 'spam'
else:
    raise Exception("Could not find label column (expected 'label' or 'spam')")

if 'text' in df.columns:
    text_col = 'text'
elif 'message' in df.columns:
    text_col = 'message'
elif 'email' in df.columns:
    text_col = 'email'
else:
    raise Exception("Could not find email text column (expected 'text', 'message', or 'email')")
spam_df = df[df[label_col].astype(str).str.lower().isin(['spam', '1'])]
print("\nTotal Emails:", len(df))
print("Total Spam Emails:", len(spam_df))

print("\nSample Spam Emails:\n")
for i, email in enumerate(spam_df[text_col].head(10), 1):
    print(f"{i}. {email}\n")

spam_df.to_csv('spam_emails_only.csv', index=False)
print("\nAll spam emails saved to 'spam_emails_only.csv'")
```
